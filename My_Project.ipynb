```python
#Importing Libraries
import pandas as pd
import numpy as np
```

```python
#Read CSV file/ Importing dataset
df=pd.read_csv('social media influencers-youtube.csv')
```

```python
#Chechking the data
#df.head()
```

```python
#Chechking the data
#df.tail()
```

```python
#Renaming column name for better knowledge
df.rename(columns={'Views (Avg.) ==0':'Views'},inplace=True)
df.rename(columns={'Likes (Avg.)':'Likes'},inplace=True)
df.rename(columns={'Comments (Avg.)':'Comments'},inplace=True)
df.rename(columns={'Followers':'Subscriber'},inplace=True)
df.rename(columns={'youtuber name':'Youtuber name'},inplace=True)
```

```python
df
```

```python
#Drop unnecessary column
df.drop(columns=['Category-2'],inplace=True)
df
```

```python
#Handlling null data
new_df=df.fillna({
    'Country':'Not-Reveal'
})
new_df
```

```python
#Conversion of data to get required information
def convert_into_int(value):
    if 'K' in value:
        return int(float(value.replace('K',''))*1000)
    elif 'M' in value:
        return int(float(value.replace('M',''))*1000000)
    else:
        return  int(value)
new_df['Comments']=new_df['Comments'].apply(convert_into_int)    
new_df['Likes']=new_df['Likes'].apply(convert_into_int)
new_df['Views']=new_df['Views'].apply(convert_into_int)    
new_df['Subscriber']=new_df['Subscriber'].apply(convert_into_int)
new_df
```

```python
new_df.describe()
```

```python
#Replacing null value as NaN
new_df=new_df.replace(0,np.NaN)
new_df
```

```python
#Handelling null data
new_df1=new_df.fillna({	
    'Likes':4.360563e+04,
    'Comments':2074.210000
})
```

```python
#Handelling null data
new_df1=new_df1.fillna({
    'Category':'Creator'
})
new_df1
```
